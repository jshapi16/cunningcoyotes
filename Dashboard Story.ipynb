```python
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv("Data\cunning_coyote_all_datasets.csv")
```

```python
data.head()
```

```python
x = data["CPI"]
y = data["first_percentile"]

(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x, y)
regress_values = x * slope + intercept

correlation = round(st.pearsonr(x,y)[0],2)
print(f'The correlation coefficient for weight and tumor volume is {correlation}')

# Plot the linear regression model
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

fig, ax = plt.subplots(figsize=(7,5));
ax.scatter(x,y, color="black");
ax.plot(x, regress_values, color="pink");
ax.annotate(line_eq,(18,38),fontsize=15,color="red")
ax.set(title="CPI vs First%",
      xlabel="CPI",
      ylabel= "first_percentile");
```
