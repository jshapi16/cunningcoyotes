```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import os
```

```python
cpi = "CPI_percentage_change.csv"
cpe = "CPE_percentage_change.csv"
gdp = "GDP_percentage_change.csv"
median_income = "median_household_income_percentage_change.csv"
percap_gdp = "per_capital_gdp_percentage_change.csv"
income_percentile = "income_by_percentile_adjusted_FINAL.csv"
```

```python
# Read the data
df_cpi = pd.read_csv(cpi)
df_cpe = pd.read_csv(cpe)
df_gdp = pd.read_csv(gdp)
df_median_income = pd.read_csv(median_income)
df_percap_gdp = pd.read_csv(percap_gdp)
df_income_perc = pd.read_csv(income_percentile)
```

```python
combine = [df_cpi, df_cpe, df_gdp, df_median_income, df_percap_gdp, df_income_perc]
```

```python
all_datasets = pd.concat(combine, axis=1)
```

```python
merge = partial(pd.merge, on=['Year'], how='outer')
```
