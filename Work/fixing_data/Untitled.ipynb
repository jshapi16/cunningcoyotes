```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import csv
import os
```

```python
all_data = "cunning_coyote_change.csv"
new_data = "format_income.csv"
```

```python
df_all = pd.read_csv(all_data)
```

```python
df_new = pd.read_csv(new_data)
```

```python
df_add = df_all[['Year', 'CPI', 'percentage_cpi', 'PCE_food_billions_of_dollars',
       'PCE_energy_goods_services_billions_of_dollars', 'PCE_food_percentage',
       'PCE_energy_goods_services_percentage', 'GDP_in_Billions',
       'percentage_gdp', 'median_household_income_2021',
       'percentage_median_income', 'GDP', 'percentage_percap_gdp','cpi_change', 'pce_food_change', 'pce_energy_change', 'gdp_change',
       'median_income_change', 'gdp_percap_change']]
```

```python
df_add
```

```python
combine = [df_add, df_new]
```

```python
from functools import partial, reduce
```

```python
merge = partial(pd.merge, on=['Year'], how='outer')
```

```python
all_dfs = reduce(merge, combine)
```

```python
all_dfs
```

```python
all_dfs.columns
```

```python
all_dfs = all_dfs[['Year', 'CPI', 'percentage_cpi', 'PCE_food_billions_of_dollars',
       'PCE_energy_goods_services_billions_of_dollars', 'PCE_food_percentage',
       'PCE_energy_goods_services_percentage', 'GDP_in_Billions',
       'percentage_gdp', 'median_household_income_2021',
       'percentage_median_income', 'GDP', 'percentage_percap_gdp','Lowest\nfifth',
       'Second\nfifth', 'Middle\nfifth', 'Fourth\nfifth', 'Highest\nfifth',
       'Top 5\npercent','cpi_change', 'pce_food_change', 'pce_energy_change', 'gdp_change',
       'median_income_change', 'gdp_percap_change']]
```

```python
all_dfs
```

```python
#change column names 
'first_percentile', 'second_percentile','third_percentile', 'fourth_percentile', 'top_five_percent'
'Lowest\nfifth','Second\nfifth', 'Middle\nfifth', 'Fourth\nfifth', 'Highest\nfifth','Top 5\npercent'
```

```python
all_dfs = all_dfs.rename(columns={'Lowest\nfifth': 'first_percentile', 'Second\nfifth': 'second_percentile', 'Middle\nfifth':'third_percentile', 'Fourth\nfifth':'fourth_percentile', 'Highest\nfifth': 'fifth_percentile', 'Top 5\npercent':'top_five_percent'})
```

```python
all_dfs
```

```python
all_dfs = all_dfs.set_index("Year")
all_dfs
```
